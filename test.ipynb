```python
from sensiml import Client

client = Client()
client.project="ml_demo_one_channel"

client.pipeline="sixth_trial"

client.pipeline.reset()

client.pipeline.set_input_query("all_data_2", use_session_preprocessor=False)


client.pipeline.add_augmentation(
   [
       {
           "name": "Scale Amplitude",
           "params": {
               "scale_range": [0.1, 1.5],   # augmenting by multiplying a segment by a factor chosen from this range
               "fraction": 1,
           },
       },

   ], overwrite=False,
)

client.pipeline.add_augmentation(
   [
       {
           "name": "Time Shift",
           "params": {
               "shift_range": [-2000, 2000],
               "averaging_window_size": 200,
               "fraction": 1,
               "filter": {},
               "replace": False,  # default is False
           },
       },
   ], overwrite=False,
)


client.pipeline.add_augmentation(
       [
           {
               "name": "Add Noise",
               "params": {
                   "background_scale_range": [50, 400],    # set the range of noise level based on your signal level and desired signal to noise ratio
                   "fraction": 1,
                   "noise_types": ["white", "pink", "violet", "blue"],     # possible options: ["white", "pink", "violet", "blue"]
                   "replace": True,
                   "filter": {},
           },
       },

   ], overwrite=False,
)    

client.pipeline.add_transform("Windowing", params={"window_size":480,
                                "delta":320})


client.pipeline.add_transform("Cascade Windowing", params={"window_size":480,
                                "delta":320})

#client.pipeline.get_results()


#client.pipeline.add_feature_generator([{'name':'MFCC V2', 'params':{"columns":['channel_0']}}])

#client.pipeline.add_transform("Min Max Scale", params={"min_bound":-128,
#                               "max_bound":127,
#                               "pad":0,
#                               "feature_min_max_defaults":{'maximum': 127, 'minimum': -128}})

#client.pipeline.add_transform("Feature Cascade", params={"num_cascades":49,
#                               "slide":True,
#                               "training_slide":True,
#                               "training_delta":1})

# stop the old pipeline if it has the same name
client.pipeline.stop_pipeline()

# execute the pipeline on the SensiML server
results, stats = client.pipeline.execute()

### End Piepline
```
